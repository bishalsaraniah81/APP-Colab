2. Draw a circle using turtle package

```python
t.initializeTurtle()
t.speed(100)
for _ in range(360):
  t.forward(1)
  t.left(1)
```

3. Draw Alphabets using turtle package

```python
t.initializeTurtle()

t.left(90)
t.forward(100)
t.right(90)
t.circle(-50, 180)
t.right(180)
t.circle(-50, 180)
t.right(90)
t.forward(100)

# hide the turtle cursor
t.hideturtle()
```

4. Draw a flower using Turtle package

```python
t.initializeTurtle()

t.speed(100)
def flower():
    for i in range(150):
      t.circle(190-i,90)
      t.left(90)
      t.circle(190-i,90)
      t.left(18)
    return flower
flower()
```

5. Draw a spiral square inside out

```python
t.initializeTurtle()
t.bgcolor("white")
t.pensize(1)
t.speed(10)
for i in range(180):
   t.pencolor("black")
   t.forward(i)
   t.left(90)
```

6. Draw a spiral hexagon

```python
t.initializeTurtle()
t.bgcolor("white")
t.pensize(5)
t.speed(10)
for i in range(10,180,2):
   t.pencolor("black")
   t.forward(i)
   t.left(59.3)
```
