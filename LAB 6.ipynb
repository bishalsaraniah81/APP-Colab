# STRUCTURED QUERY LANGUAGE

1. Create a table with name CRICKETERS with the column names

a. First name

b. Last name

c. Age

d. Place of birth

e. Country

```python
import sqlite3
conn = sqlite3.connect('Cricket.db')
con=conn.cursor()
con.execute("""CREATE TABLE CUSTOMERS
  (Firstname VARCHAR(50),Lastname VARCHAR(50),Age INT,Placeofbirth VARCHAR(50),Country VARCHAR5(0))""")
```

3. List all the rows

```python
con.execute("SELECt * from CUSTOMERS")
rows=con.fetchall()
for row in rows:
  print(row)
```

4. Retrieve the rows whose age is greater than 35.`

```python
con.execute("SELECt * from CUSTOMERS WHERE Age>35")
rows=con.fetchall()
for row in rows:
  print(row)
```

5. List all the rows in ascending order (First name)

```python
con.execute("SELECt * from CUSTOMERS ORDER BY Firstname")
rows=con.fetchall()
for row in rows:
  print(row)
```
