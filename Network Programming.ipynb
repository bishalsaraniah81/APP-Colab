```python
#TCP Client
import socket
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
host= socket.gethostname()
host_name = socket.gethostbyname (host)
port = 12345
client.connect((host, port))
print("Host name:", host_name)
data = client.recv(1024)
print("Message from server: ", data.decode())
reply = "Your welcome!" 
client.send(reply.encode())
client.close()
```

```python
#TCP Server
import socket
#No.1 socket()
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
host = socket.gethostname()
host_name = socket.gethostbyname (host)
port = 12345
print("Host name:", host_name)
#No.2 bind()
server_socket.bind ((host, port)) 
print("Socket binded to %s" %(port))
#No.3 listen() server_socket.listen(5) print("Server is listening")
#No.4 accept, send receive
while True:
  client_socket, addr = server_socket.accept() 
  print("Got connection from", addr)
  msg = "Thank u for connecting" 
  en_msg=msg.encode() 
  client_socket.send(en_msg)
  data= client_socket.recv(1024)
  print("Message from client: ", data.decode())
client_socket.close()
```

```python
#UDP CLIENT
import socket
client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
host = socket.gethostname()
port = 12345
msg = "Hello world"
en_msg = msg.encode()
client.sendto(en_msg, (host, port))
```

```python
#UDP SERVER
import socket
#No.1 socket()
server_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
host
socket.gethostname()
port = 12345
#No.2 bind()
server_socket.bind ((host, port)) 
print("Socket binded to %s" % (port))
#No.3 sendto / receive from
while True:
  msg,addr = server_socket.recvfrom(1024) 
  de_msg = msg.decode()
  print("Message from client: ", de_msg, "from", addr)
```
