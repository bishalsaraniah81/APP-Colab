```python
# PROGRAM 2
# Python Program to create a Fibonacci Series

def fibonacci(n):
  if n<0:
    print("Invalid")
  elif n==0:
    return 0
  elif n==1 or n==2:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
  
a = int(input("Enter the number : "))
print("Fibonacci of ",a,"is ",fibonacci(a))
```

PROGRAM 5

```python
a=[]
n=int(input("Enter no of elements : "))
for i in range(0,n):
  val=int(input())
  a.append(val)
for i in range(n-1):
  for j in range(n-i-1):
    if a[j]>a[j+1]:
      a[j],a[j+1]=a[j+1],a[j]
print(a)
```

```python
def LengthUsingForLoop(str):
  counter=0
  for i in str:
    counter += 1
  return counter

def LengthUsingSlicing(str):
  counter = 0
  while str[counter:]:
    counter +=1
  return counter

def LengthUsingJoinAndCount(str):
  if not str:
    return 0
  else:
    random_str="pyqy"
    return((random_str).join(str)).count(random_str)+1

a =input("Enter string: ")
print("Using in-build length function:",len(a))
print("Using for-loop:",LengthUsingForLoop(a))
print("Using slicing and while-loop",LengthUsingSlicing(a))
print("Using join and count:", LengthUsingJoinAndCount(a))
```

PROGRAM 8

```python
def LengthWithoutSpace(str):
  len=0
  for i in str:
    if i!=" ":
      len+=1
  return len

s = input("Enter the string: ")
print("Length with spaces: ",len(s))
print("Length without spaces: ",LengthWithoutSpace(s))
```
