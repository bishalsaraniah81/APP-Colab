PROGRAM 3C - AGE CALCULATOR

```python
from datetime import date
 
def calculateAge(born):
    today = date.today()
    try:
        birthday = born.replace(year = today.year)

    except ValueError:
        birthday = born.replace(year = today.year,
                  month = born.month + 1, day = 1)
 
    if birthday > today:
        return today.year - born.year - 1
    else:
        return today.year - born.year
         

print(calculateAge(date(2004, 3, 3)), "years")
```

PROGRAM 3D - EVEN OR ODD USING CLASS AND OBJECT

```python
class Test:

    def check(self, a):
        if a % 2 == 0:
            print("Number is even:", a)
        else:
            print("Number is odd:", a)


a = int(input("Enter a number:"))
obj = Test()
obj.check(a)
```

PROGRAM 3E - SUM OF 2 NUMBERS USING INHERITANCE

```python
class First:
    def input1(self):
        self.x = int(input("Enter first number:"))

class Second:
    def input2(self):
        self.y = int(input("Enter second number:"))

class Third(First, Second):
    def add(self):
        super().input1()
        super().input2()
        self.z = self.x + self.y
        print("Sum is:", self.z)

obj = Third()
obj.add()
```

PROGRAM 3F - STUDENT DETAILS USING SINGLE INHERITANCE

```python
class Student_Details:
    def __init__(self):
        self.name = input("Enter Your Name: ")
        self.cname = input("Enter Your College Name: ")
        self.roll = int(input("Enter Your Roll Number: "))
class Temp(Student_Details):
    def display(self):
        print(" Student Details are ")
        print("Name: ", self.name)
        print("College Name:", self.cname)
        print("Roll number:", self.roll)
obj1 = Temp()
obj1.display()
```

PROGRAM 3G - THREE NUMBERS USING SINGLE INHERITANCE

```python
class First:
    def input(self):
        self.a = int(input(" "))
        self.b = int(input(" "))
        self.c = int(input(" "))

class Second(First):
    def findsum(self):
        self.result = self.a + self.b + self.c

    def display(self):
        print("Sum of three numbers:", self.result)

obj = Second()
obj.input()
obj.findsum()
obj.display()
```
