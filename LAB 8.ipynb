#7a.THREADING CONCEPT

```python
import threading
def testThread(num):
  num=num**3
  print(num)

for i in range(6):
  t = threading.Thread(target=testThread, args=(i,))
  t.start()
```

#7b. MULTIPROCESSING CONCEPT

```python
import multiprocessing
def spawn():
  print("Hello Mic Check")

if __name__=='__main__':
  for i in range(3):
    print("First Loop")
    p = multiprocessing.Process(target=spawn)
    p.start()
    print("Waiting")
  for i in range(3):
    print("Second Loop")
    p = multiprocessing.Process(target=spawn)
    p.start()
    p.join()
```

#7c. PARALLEL PROGRAMMING USING PROCESS CLASS

```python
import time
from multiprocessing import Process

def test(fname):
  f = open(fname,"w")
  f.write("hi")
  f.write("hi")
  f.write("hi")
  f.write("hi")
  f.close()

if __name__ == '__main__':
  start_time = time.time()
  processlist = []
  p1 = Process(target = test, args = ("Sample1.txt",))
  p2 = Process(target = test, args = ("Sample2.txt",))
  p1.start()
  p2.start()
  p1.join()
  p2.join()
  end_time = time.time()
  print(f"Time taken {end_time - start_time} seconds")
```

#7d. PARALLEL PROGRAMMING USING POOL CLASS

```python
import time
from multiprocessing import Pool

def test(fname):
  f = open (fname,"w")
  f.write("Hi Kamal")
  f.write("Hi Kamal")
  f.write("Hi Kamal")
  f.write("Hi Kamal")
  f.close()

if __name__=="__main__":
  starttime= time.time()
  pool= Pool()
  a = pool.apply_async(test, args=("sample1.txt", ))
  b = pool.apply_async(test, args=("sample2.txt", ))
  a.wait()
  b.wait()
  endtime= time.time()
  print(f"Time Taken {endtime-starttime} Seconds")
```
